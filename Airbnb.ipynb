```python
import pandas as pd

# Load the Airbnb data (adjust the filename/path as needed)
df = pd.read_csv("airbnb_listings.csv")

# Count the number of listings by room type
room_counts = df['room_type'].value_counts()
print("Room Type Counts:")
print(room_counts)

# Sum up listings that correspond to Airbnb's initial idea (i.e. rooms inside someone's home)
initial_listings = df[df['room_type'].isin(['Private room', 'Shared room'])].shape[0]

# Total listings in the dataset
total_listings = df.shape[0]

# Calculate the proportion of 'room' listings
initial_prop = initial_listings / total_listings
print(f"\nPercentage of listings that are rooms or shared rooms: {initial_prop:.2%}")

# Decide whether Airbnb is still mainly about rooms or shared spaces
if initial_prop < 0.5:
    print("\nNo, Airbnb's offerings have expanded beyond just rooms or shared spaces in someone's home.")
else:
    print("\nYes, Airbnb is still mainly offering rooms or shared spaces.")
```

```python
import pandas as pd
import tkinter as tk
from tkinter import filedialog

# Set up the tkinter root and hide the main window
root = tk.Tk()
root.withdraw()

# Open a file dialog and let the user select a CSV file
file_path = filedialog.askopenfilename(title="Select the Airbnb CSV file",
                                       filetypes=[("CSV files", "*.csv")])

# Check if a file was selected
if file_path:
    df = pd.read_csv(file_path)
    print("File uploaded successfully! Here's a preview:")
    print(df.head())
else:
    print("No file was selected.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "listings_detailed.csv"  # Update this with the correct path if needed
df = pd.read_csv(file_path)

# Analyze the distribution of room types
room_type_distribution = df['room_type'].value_counts(normalize=True) * 100

# Print the distribution
print("Room Type Distribution:")
print(room_type_distribution)

# Plot the distribution of room types
plt.figure(figsize=(8, 6))
room_type_distribution.plot(kind='bar', color=['blue', 'green', 'red', 'orange'])
plt.title("Distribution of Room Types on Airbnb")
plt.xlabel("Room Type")
plt.ylabel("Percentage of Listings")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show theÂ plot
plt.show()
```
